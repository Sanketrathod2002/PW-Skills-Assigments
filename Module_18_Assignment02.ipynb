#**Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of Matplotlib.**

Matplotlib is a plotting library in Python used to create static, animated, and interactive visualizations in Python. It provides a wide variety of plots and charts for visualizing data in a clear and effective manner. Some of the plots that can be plotted using the Pyplot module of Matplotlib include:

1. Line plot
2. Scatter plot
3. Bar plot
4. Histogram
5. Pie chart

#**Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

# Plot scatter plot
plt.scatter(x, y)
plt.title("Scatter Plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

#**Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:**

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for line plots
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create subplots and plot line plots
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line Plot 1")

plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line Plot 2")

plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line Plot 3")

plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line Plot 4")

plt.tight_layout()
plt.show()
```

#**Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for bar plot
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Plot bar plot
plt.figure(figsize=(10, 6))
plt.bar(company, profit)
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()

# Plot horizontal bar plot
plt.figure(figsize=(10, 6))
plt.barh(company, profit)
plt.title("Horizontal Bar Plot")
plt.xlabel("Profit")
plt.ylabel("Company")
plt.show()
```

#**Q5: What is a box plot? Why is it used? Using the following data plot a box plot.**

```python
import matplotlib.pyplot as plt
import numpy as np

# Data for box plot
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)

# Plot box plot
plt.figure(figsize=(8, 6))
plt.boxplot([box1, box2])
plt.title("Box Plot")
plt.xlabel("Data")
plt.ylabel("Values")
plt.xticks([1, 2], ['Box 1', 'Box 2'])
plt.show()
```
