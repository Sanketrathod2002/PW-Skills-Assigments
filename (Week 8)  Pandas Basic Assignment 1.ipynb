```python
'''Q4. What is ‘DataFrame’ in pandas and how is it different from pandas.series? Explain with an example.'''
#Ans :
'''
A Series in Pandas is like a single column of data, similar to a list or an array.
A DataFrame in Pandas is like a table or a spreadsheet, consisting of multiple columns (Series) 
that are related to each other, creating a structured and organized way to store and analyze data.
'''
#Example :
import pandas as pd

# Series
fruits_series = pd.Series(['Apple', 'Banana', 'Cherry', 'Date'], name='Fruit')

# DataFrame
data = {
    'Fruit': ['Apple', 'Banana', 'Cherry', 'Date'],
    'Color': ['Red', 'Yellow', 'Red', 'Brown'],
    'Taste': ['Sweet', 'Sweet', 'Sweet', 'Sweet'],
}

fruits_df = pd.DataFrame(data)

# Printing the Series and DataFrame
print("Series (Single Column):")
print(fruits_series)

print("\nDataFrame (Table with Multiple Columns):")
print(fruits_df)
```

```python
'''Q7. Create a DataFrame using multiple Series. Explain with an example.'''

#Ans :
import pandas as pd

# Creating two Pandas Series
names = pd.Series(['Sanket', 'Kartik', 'Mayur'])
ages = pd.Series([21, 20, 18])

# Creating a DataFrame by combining the Series into a dictionary
data = {'Name': names, 'Age': ages}
df = pd.DataFrame(data)

# Printing the DataFrame
print(df)
```
