#**Q1: What are the Probability Mass Function (PMF) and Probability Density Function (PDF)? Explain with an example.**

- Probability Mass Function (PMF): PMF is a function that gives the probability that a discrete random variable is exactly equal to some value. It maps each possible value of the random variable to its probability of occurring.

- Probability Density Function (PDF): PDF is a function that describes the likelihood of a continuous random variable falling within a particular range of values. It gives the probability of a continuous random variable taking on a specific value.

Example:
For a fair six-sided die, the PMF would be:
PMF(1) = 1/6, PMF(2) = 1/6, PMF(3) = 1/6, PMF(4) = 1/6, PMF(5) = 1/6, PMF(6) = 1/6

For a continuous random variable such as the height of individuals in a population, the PDF would describe the likelihood of observing a particular height value within a certain range.

#**Q2: What is Cumulative Density Function (CDF)? Explain with an example. Why CDF is used?**

- Cumulative Density Function (CDF): CDF gives the probability that a random variable takes on a value less than or equal to a given value. It accumulates the probabilities as you move through the range of possible values of the random variable.

Example:
For a fair six-sided die, the CDF would be:
CDF(1) = 1/6, CDF(2) = 2/6, CDF(3) = 3/6, CDF(4) = 4/6, CDF(5) = 5/6, CDF(6) = 1

CDF is used to determine the probability of a random variable falling within a specific range or below a certain value.

#**Q3: What are some examples of situations where the normal distribution might be used as a model? Explain how the parameters of the normal distribution relate to the shape of the distribution.**

The normal distribution might be used as a model in situations such as:
- Heights or weights of individuals in a population.
- Test scores of students in a large class.
- Errors in measurement instruments.

The parameters of the normal distribution are the mean (μ) and the standard deviation (σ). The mean determines the center of the distribution, while the standard deviation determines the spread or dispersion of the distribution. A larger standard deviation results in a wider and flatter distribution.

The normal distribution is important because:
- It is widely used in statistics and probability theory due to its simplicity and mathematical properties.
- Many natural phenomena and human characteristics tend to follow a normal distribution.

Examples of real-life situations where the normal distribution applies include:
- Heights of individuals in a population.
- IQ scores of a large group of people.
- Blood pressure measurements in a healthy adult population.

#**Q5: What is Bernoulli Distribution? Give an Example. What is the difference between Bernoulli Distribution and Binomial Distribution?**

- Bernoulli Distribution: Bernoulli distribution models the probability of success or failure in a single trial of a binary experiment.

Example: Flipping a fair coin, where success (1) is getting heads and failure (0) is getting tails.

- Difference between Bernoulli Distribution and Binomial Distribution:
  - Bernoulli distribution has only one trial, whereas binomial distribution has multiple independent trials.
  - Binomial distribution is the sum of multiple independent Bernoulli trials.

#**Q7: Explain uniform Distribution with an example.**

Uniform distribution is a probability distribution where all outcomes are equally likely. In other words, each value in the range has the same probability of occurring.

Example: Rolling a fair six-sided die, where each side has a probability of 1/6 of being rolled.

#**Q8: What is the z score? State the importance of the z score.**

The z-score (or standard score) measures the number of standard deviations a data point is from the mean of the dataset. It allows for standardized comparison of data points from different normal distributions. The importance of the z-score lies in its ability to quantify how extreme or unusual a data point is relative to the mean of the distribution.

#**Q9: What is Central Limit Theorem? State the significance of the Central Limit Theorem.**

Central Limit Theorem states that the sampling distribution of the sample mean approaches a normal distribution as the sample size increases, regardless of the shape of the population distribution. It is significant because it allows us to make inferences about population parameters based on sample statistics, even when the population distribution is unknown or non-normal.

---
#**Q10: State the assumptions of the Central Limit Theorem.**

The assumptions of the Central Limit Theorem are:
- The sample is randomly selected from the population.
- The sample size is sufficiently large (usually n ≥ 30).
- The sample observations are independent of each other.
- The population standard deviation is known (for z-test) or the sample size is large enough for the sample standard deviation to approximate the population standard deviation (for t-test).
