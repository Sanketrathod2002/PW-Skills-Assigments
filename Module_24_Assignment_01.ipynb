**Q1. Load the flight price dataset and examine its dimensions. How many rows and columns does the dataset have?**

```python
import pandas as pd

flight_data = pd.read_excel("/content/flight_price.xlsx")

print("Number of rows and columns:", flight_data.shape)
```

**Q2. What is the distribution of flight prices in the dataset? Create a histogram to visualize the distribution.**

```python
import matplotlib.pyplot as plt

# Plot histogram of flight prices
plt.figure(figsize=(10, 6))
plt.hist(flight_data['Price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Flight Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

**Q4. How does the price of flights vary by airline? Create a boxplot to compare the prices of different airlines.**

```python
# Create a boxplot to compare prices by airline
plt.figure(figsize=(12, 8))
plt.boxplot([flight_data[flight_data['Airline'] == airline]['Price'] for airline in flight_data['Airline'].unique()],
            labels=flight_data['Airline'].unique())
plt.xticks(rotation=90)
plt.title('Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

**Q6. You are working for a travel agency, and your boss has asked you to analyze the Flight Price dataset to identify the peak travel season. What features would you analyze to identify the peak season, and how would you present your findings to your boss?**

To identify the peak travel season, I would analyze the 'Date_of_Journey' feature, which indicates the date of the flight journey. Specifically, I would extract the month from the date and analyze the distribution of flight prices by month. This would help in identifying months with higher average flight prices, indicating peak travel seasons.

To present the findings to my boss, I would create a bar plot or line plot showing the average flight prices by month. This visualization would clearly highlight the peak travel seasons when flight prices are higher.

```python
# Extract month from the 'Date_of_Journey' column
flight_data['Month'] = pd.to_datetime(flight_data['Date_of_Journey']).dt.month

# Calculate the average flight prices by month
average_prices_by_month = flight_data.groupby('Month')['Price'].mean()

# Plot average flight prices by month
plt.figure(figsize=(10, 6))
average_prices_by_month.plot(kind='bar', color='skyblue')
plt.title('Average Flight Prices by Month')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

**Q7. You are a data analyst for a flight booking website, and you have been asked to analyze the Flight Price dataset to identify any trends in flight prices. What features would you analyze to identify these trends, and what visualizations would you use to present your findings to your team?**

To identify trends in flight prices, I would analyze the following features:

Date of Journey: Analyzing how flight prices vary over different dates, months, or seasons.
Airline: Examining how prices vary across different airlines.
Source and Destination: Investigating if flight prices differ based on the source and destination cities.
Total Stops: Exploring how the number of stops affects flight prices.
Duration: Investigating if the duration of the flight impacts the prices.
To present my findings to the team, I would use various visualizations such as line plots, bar plots, boxplots, and scatter plots. These visualizations would help in clearly illustrating the trends and patterns observed in flight prices based on different features.

```python
# Analyzing flight prices based on different features
# 1. Date of Journey
flight_data['Day_of_Journey'] = pd.to_datetime(flight_data['Date_of_Journey']).dt.day
flight_data['Month_of_Journey'] = pd.to_datetime(flight_data['Date_of_Journey']).dt.month

# Plotting average flight prices by day and month
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
flight_data.groupby('Day_of_Journey')['Price'].mean().plot(kind='line', color='skyblue')
plt.title('Average Flight Prices by Day of Journey')
plt.xlabel('Day')
plt.ylabel('Average Price')
plt.grid(True)

plt.subplot(1, 2, 2)
flight_data.groupby('Month_of_Journey')['Price'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Flight Prices by Month of Journey')
plt.xlabel('Month')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Analyzing flight prices based on different features
# 2. Airline
plt.figure(figsize=(12, 6))
flight_data.groupby('Airline')['Price'].mean().sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Flight Prices by Airline')
plt.xlabel('Airline')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Analyzing flight prices based on different features
# 3. Source and Destination
plt.figure(figsize=(12, 6))
flight_data.groupby(['Source', 'Destination'])['Price'].mean().sort_values().plot(kind='bar', color='skyblue')
plt.title('Average Flight Prices by Source and Destination')
plt.xlabel('Source - Destination')
plt.ylabel('Average Price')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Analyzing flight prices based on different features
# 4. Total Stops
plt.figure(figsize=(10, 6))
flight_data.groupby('Total_Stops')['Price'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Flight Prices by Total Stops')
plt.xlabel('Total Stops')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```

```python
# Analyzing flight prices based on different features
# 5. Duration
# Convert duration to minutes for better analysis
def convert_duration_to_minutes(duration):
    duration_parts = duration.split(' ')
    if len(duration_parts) == 2:  # Check if duration is in the expected format
        hours = int(duration_parts[0][:-1])
        minutes = int(duration_parts[1][:-1])
        return hours * 60 + minutes
    else:
        return 0  # Return 0 if the duration format is unexpected

flight_data['Duration_minutes'] = flight_data['Duration'].apply(convert_duration_to_minutes)

plt.figure(figsize=(10, 6))
plt.scatter(flight_data['Duration_minutes'], flight_data['Price'], color='skyblue', alpha=0.5)
plt.title('Flight Prices vs. Duration')
plt.xlabel('Duration (minutes)')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```

**Q13. How does the type of app affect its price? Create a bar chart to compare average prices by app type.**

```python
import pandas as pd
import matplotlib.pyplot as plt

google_playstore_data = pd.read_csv("https://raw.githubusercontent.com/krishnaik06/playstore-Dataset/main/googleplaystore.csv")

google_playstore_data = google_playstore_data[google_playstore_data['Price'].apply(lambda x: str(x).replace('.', '').isdigit())]

google_playstore_data['Price'] = pd.to_numeric(google_playstore_data['Price'])

average_prices_by_type = google_playstore_data.groupby('Type')['Price'].mean()

plt.figure(figsize=(8, 6))
average_prices_by_type.plot(kind='bar', color=['skyblue', 'orange'])
plt.title('Average Prices by App Type')
plt.xlabel('App Type')
plt.ylabel('Average Price')
plt.xticks(rotation=0)
plt.show()
```
